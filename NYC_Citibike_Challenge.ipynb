```python
# 3. Convert the 'tripduration' column to datetime datatype.
# create a new column that contains the conversion keep the "tripduration" column to int64 type
citibike_tripdata['tripdatetime'] = ""
citibike_tripdata['tripdatetime'] = pd.to_datetime(citibike_tripdata['tripduration'], unit='m')
citibike_tripdata['tripdatetime']
```
